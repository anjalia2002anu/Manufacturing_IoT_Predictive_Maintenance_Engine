```python
import json
import requests

with open("sample_input.json", "r") as f: #Reading data from a JSON file & Converts it into a Python dictionary
    sample_input = json.load(f)

url = "https://dispensational-doily-jonelle.ngrok-free.dev/predict" #This is the public API endpoint, Provided by ngrok

response = requests.post(url, json=sample_input) #Sending POST request

print(response.status_code)
print(response.json()) #Reading API response
```
